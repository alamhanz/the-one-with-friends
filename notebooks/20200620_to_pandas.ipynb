## Implement all

```python
df_trcpt_all=pd.DataFrame()
for tx in os.listdir(PATH_RES):
    print(tx)
    if 'pkl' in tx:
        summary=joblib.load(PATH_RES+tx)
        transcript=summary['dialogue']
        
        df_transcript0=pd.DataFrame(transcript,columns=['ori'])
        df_transcript0['dialogue_ids']=df_transcript0['ori'].str.split('~').apply(lambda x: x[0])
        df_transcript0['ori2']=df_transcript0['ori'].str.split('~').apply(lambda x: x[1])

        df_transcript0['scenes_places']=df_transcript0['ori2'].str.split('-').apply(lambda x: x[0])
        df_transcript0['ori3']=df_transcript0['ori2'].str.split('-').apply(lambda x: x[1])

        df_transcript0['character']=df_transcript0['ori3'].str.split(':').apply(lambda x: x[0])
        df_transcript0['dialogue']=df_transcript0['ori3'].str.split(':').apply(lambda x: x[1])

        df_transcript0[['ori','dialogue_ids','scenes_places','character','dialogue']]
        
        df_transcript0['eps']=tx[:-4]
        df_trcpt_all=df_trcpt_all.append(df_transcript0[['eps','dialogue_ids','scenes_places','character','dialogue']])
```
