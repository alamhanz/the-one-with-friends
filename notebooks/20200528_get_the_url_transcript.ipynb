```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import lxml.html as lh
import pandas as pd
from urllib.request import urlopen
# from selenium import webdriver
import time
import numpy as np
import random
import os
import re
# import nltk
```

```python
PATH_SOURCE='../data/friends-transcripts/'
```

# Source

```python
html_source = urlopen('https://fangj.github.io/friends/').read().decode("utf-8")
doc = lh.fromstring(html_source)
```

```python
all_links=doc.xpath('//li//a')
```

# All Seasons

```python
all_episodes=all_links[0].xpath("//li//a/@href")
all_episodes=['https://fangj.github.io/friends/'+i for i in all_episodes]
# all_episodes
```

```python
one_eps=all_episodes[0]
# 'outtakes'
```

```python
rr='01'
for one_eps in all_episodes:
    if 'outtakes' not in one_eps:
        html_source0=urlopen(one_eps).read().decode("utf-8",errors='ignore')    
        notag = re.sub("<.*?>", " ", html_source0)

        notag = notag.replace('&quot;','')
        notag = notag.replace('&nbsp;','')
        notag = notag.replace('&amp;','')

        notag = notag.replace('\n\n','\n')
        notag = notag.replace('\n','~')
        notag = re.sub("<.*?>", " ", notag)
        notag = notag.replace('~','\n')

        fn_txt=one_eps.split('/')[-1].split('.')[0]
        file_name=PATH_SOURCE+fn_txt+".txt"
        file_script = open(file_name,"w")
        file_script.write(notag)
        file_script.close()

        if fn_txt[:2]==rr:
            print(rr)
        else:
            rr=fn_txt[:2]
            print(rr)
```
