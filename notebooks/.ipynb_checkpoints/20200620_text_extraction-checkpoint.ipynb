## Run All

```python
for eps in os.listdir(PATH_DATA):
    if '.txt' in eps:
        print(eps)
        transcript=open(PATH_DATA+eps)
        transcript_lines0=transcript.readlines()
        transcript_lines0=[i.lower().strip().replace('&#146;',"'").replace('&#133;','...')
                          .replace('&#151;','--').replace('&#145;',"")
                          .replace('&#147;','"').replace('&#148;','"')
                          .replace('&#150;',"-").replace('&#8216;',"'")
                          .replace('&#8217;',"'").replace('&#8220;','"')
                          .replace('&#8221;','"').replace('&#8230;','...') for i in transcript_lines0]
        
        transcript_lines=[i.replace('.',',') if i.startswith('[scene') else i for i in transcript_lines0]
        transcript_lines=[i.replace('-',' ') for i in transcript_lines]
        transcript_lines=preprocessing(transcript_lines)
        
        
        res_extract=extraction(transcript_lines)
        joblib.dump(res_extract,PATH_RES+eps[:-4]+'.pkl')
```
